```python
import pandas as pd
import json

# ── Load data ────────────────────────────────────────
sales_df = pd.read_csv("../data/raw/sales_transactions.csv")
with open("../data/raw/customer_profiles.json") as f:
    customers = json.load(f)
cust_df = pd.DataFrame(customers)

# ── Sales overview ───────────────────────────────────
print("Shape:", sales_df.shape)
print("\nData types:\n", sales_df.dtypes)
print("\nNull counts:\n", sales_df.isnull().sum())
print("\nRevenue by region:\n",
      sales_df.groupby("region")["total_revenue"].sum().sort_values(ascending=False))
print("\nTop 3 products:\n",
      sales_df.groupby("product_name")["total_revenue"].sum().nlargest(3))

# ── Customer overview ────────────────────────────────
print("\nCustomer segments:\n", cust_df["segment"].value_counts())
print("\nCities:\n", cust_df["city"].value_counts().head(5))
```
